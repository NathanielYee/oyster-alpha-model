```python
import requests
import pandas as pd

# Get data from DexScreener API
def pair_data(tokenAddresses: str):
    response = requests.get(
        url=f"https://api.dexscreener.com/latest/dex/tokens/{tokenAddresses}"
    )
    
    if response.status_code == 200:
        return response.json()  # Return the JSON data as a dictionary
    else:
        raise Exception(f"Failed to fetch data: {response.status_code}, {response.text}")

# Call pair_data and process the response
data = pair_data(tokenAddresses="0x8B0E6f19Ee57089F7649A455D89D7bC6314D04e8")

# Extract the 'pairs' field (assuming this is where the relevant data is stored)
pairs_data = data.get("pairs", [])

# Flatten the data into a DataFrame
data_df = pd.json_normalize(pairs_data)

# Display the DataFrame
data_df
```

```python
# Calculate p/l of the token
trade_price = 40.00
data_df['priceUsd'] = data_df['priceUsd'].astype(float)
data_df['p/l%'] = (data_df['priceUsd'] - trade_price) / data_df['priceUsd']
data_df['p/l%']
```
